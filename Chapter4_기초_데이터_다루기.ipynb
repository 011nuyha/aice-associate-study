# Section 04 데이터프레임 병합하기

## 1 concat()을 이용해 병합하기
concat() 메소드(concatenate)는 **'데이터 타입이 동일한 2개 이상의 데이터프레임'**을 병합할 때 사용한다.

```python
import pandas as pd
import numpy as np
```

### 1) 칼럼명이 같은 경우
concate() 메소드의 파라미터 ignore_index와 axis를 이용한다.
각 파라미터의 디폴트값을 통해, concat()은 기본적으로 행 레벨에서 인덱스를 유지한 채 병합한다는 것을 알 수 있다.
1. ignore_index
*   False(default) : 기존 인덱스 유지
*   True : 기존 인덱스를 무시하고 재배열
2. axis
*   0(default) : 행 레벨(세로)로 병합
*   1 : 칼럼 레벨(가로)로 병합

```python
df1 = pd.DataFrame({
    'col1' : ['사과','배','감','수박','메론'],
    'col2' : [500, 1000, 2500, 5000,3000]
    }, index=[0,1,2,3,4])
df2 = pd.DataFrame({
    'col1' : ['수박','메론','딸기','키위','오렌지'],
    'col2' : [5000,3000,1000,600,700]
    }, index=[3,4,5,6,7])
```

```python
df1, df2
```

```python
pd.concat([df1, df2]) # ignore_index 파라미터는 False가 디폴트값이다. 기존의 인덱스를 유지한다.
# pd.concat([df1, df2], ignore_index=False, axis=0)와 동일
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1) # 칼럼 레벨로 병합하는 경우, 서로 다른 인덱스에서 결측치가 발생한다.
```

### 2) 칼럼명이 다른 경우
concate() 메소드의 파라미터 join을 이용한다.
*   join=outer : 두 데이터프레임의 각 칼럼들의 합집합
*   join=inner : 두 데이터프레임의 각 칼럼들의 교집합

```python
df3 = pd.DataFrame({
    'item':['item0','item1','item2', 'item3'],
    'count':['count0','count1','count2','count3'],
    'price':['price0','price1','price2','price3']
    }, index = [0,1,2,3])
df4 = pd.DataFrame({
    'item':['item2','item3','item4', 'item5'],
    'count':['count2','count3','count4','count5'],
    'price':['price2','price3','price4','price5'],
    'var':['var2','var3','var4','var5']
    }, index = [2,3,4,5])
```

```python
df3, df4
```

```python
pd.concat([df3, df4], join='outer') # 칼럼들의 합집합
```

```python
pd.concat([df3, df4], join='inner') # 칼럼들의 교집합
```

### 3) 인덱스가 중복인 경우
concate() 메소드의 파라미터 verify_integrity를 이용한다. 데이터프레임에서 인덱스의 중복 여부를 점검할 수 있다.

```python
df5 = pd.DataFrame({
    'A':['A0', 'A1', 'A2'],
    'B':['B0', 'B1', 'B2'],
    'C':['C0', 'C1', 'C2'],
    'D':['D0', 'D1', 'D2'],
}, index=['I0','I1','I2'])

df5
```

```python
df6 = pd.DataFrame({
    'A':['AA2', 'A3', 'A4'],
    'B':['BB2', 'B3', 'B4'],
    'C':['CC2', 'C3', 'C4'],
    'D':['DD2', 'D3', 'D4'],
}, index=['I2','I3','I4'])

df6
```

```python
pd.concat([df5, df6], verify_integrity=True) # 인덱스 중복 여부 검사
```

```python
pd.concat([df5, df6]) # 인덱스 중복 무시, 행 레벨 결합, 기존 인덱스 유지
```

## 2 merge()와 join()을 이용해 병합하기
1. merge() :특정한 key를 기준으로 데이터프레임을 병합할 수 있다.
*   on='칼럼명' : 어떤 칼럼을 기준으로 merge할지 설정
*   how='아래의 값들' : merge 방식을 설정
  * inner(default) : 기준으로 삼은 칼럼('칼럼명')에서 두 데이터프레임에 모두 존재하는 데이터만 가져온다.
  * left : 왼쪽 데이터프레임의 '칼럼명'의 값들을 기준으로 한다.
  * right : 오른쪽 데이터프레임의 '칼럼명'의 값들을 기준으로 한다.
  * outer : 두 데이터프레임의 '칼럼명'의 값들을 모두 사용한다.

```python
customer = pd.DataFrame({'customer_id' : np.arange(6),
                    'name' : ['James', 'Elly', 'Tom', 'Givert', 'Aiden', 'Brody'],
                    '나이' : [40, 20, 21, 30, 31, 18]})

orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                    'item' : ['마우스', '충전기', '이어폰', '헤드셋', '전자펜', '키보드', '전자펜', '마우스', '키보드', '케이스'],
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
```

```python
customer
```

```python
orders
```

```python
pd.merge(customer, orders, on='customer_id')
# pd.merge(customer, orders, on='customer_id', how='inner')와 동일하다.
# customer_id가 두 데이터프레임에 모두 존재하는 데이터만 가져온다. => 결측치가 발생하지 않는다.
```

```python
pd.merge(customer, orders, on='customer_id', how='left') # 왼쪽 데이터프레임을 기준으로 merge한다.
```

```python
pd.merge(customer, orders, on='customer_id', how='right') # 오른쪽 데이터프레임을 기준으로 merge한다. ⭐ 이때 칼럼의 순서는 변하지 않는다.
```

```python
pd.merge(customer, orders, on='customer_id', how='outer') # 두 데이터프레임의 'customer_id'를 모두 활용해 merge한다.
```

```python
# 인덱스를 지정하여 데이터프레임 merge하기
cust1 = customer.set_index('customer_id')
order1 = orders.set_index('customer_id')
pd.merge(cust1, order1, left_index=True, right_index=True) # on 파라미터 대신에 left_index와 right_index를 사용한다.
```
