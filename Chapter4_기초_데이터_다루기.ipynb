## 1 concat()ÏùÑ Ïù¥Ïö©Ìï¥ Î≥ëÌï©ÌïòÍ∏∞
concat() Î©îÏÜåÎìú(concatenate)Îäî **'Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏù¥ ÎèôÏùºÌïú 2Í∞ú Ïù¥ÏÉÅÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ'**ÏùÑ Î≥ëÌï©Ìï† Îïå ÏÇ¨Ïö©ÌïúÎã§.

```python
import pandas as pd
import numpy as np
```

### 1) ÏπºÎüºÎ™ÖÏù¥ Í∞ôÏùÄ Í≤ΩÏö∞
concate() Î©îÏÜåÎìúÏùò ÌååÎùºÎØ∏ÌÑ∞ ignore_indexÏôÄ axisÎ•º Ïù¥Ïö©ÌïúÎã§.
Í∞Å ÌååÎùºÎØ∏ÌÑ∞Ïùò ÎîîÌè¥Ìä∏Í∞íÏùÑ ÌÜµÌï¥, concat()ÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Ìñâ Î†àÎ≤®ÏóêÏÑú Ïù∏Îç±Ïä§Î•º Ïú†ÏßÄÌïú Ï±Ñ Î≥ëÌï©ÌïúÎã§Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.
1. ignore_index
*   False(default) : Í∏∞Ï°¥ Ïù∏Îç±Ïä§ Ïú†ÏßÄ
*   True : Í∏∞Ï°¥ Ïù∏Îç±Ïä§Î•º Î¨¥ÏãúÌïòÍ≥† Ïû¨Î∞∞Ïó¥
2. axis
*   0(default) : Ìñâ Î†àÎ≤®(ÏÑ∏Î°ú)Î°ú Î≥ëÌï©
*   1 : ÏπºÎüº Î†àÎ≤®(Í∞ÄÎ°ú)Î°ú Î≥ëÌï©

```python
df1 = pd.DataFrame({
    'col1' : ['ÏÇ¨Í≥º','Î∞∞','Í∞ê','ÏàòÎ∞ï','Î©îÎ°†'],
    'col2' : [500, 1000, 2500, 5000,3000]
    }, index=[0,1,2,3,4])
df2 = pd.DataFrame({
    'col1' : ['ÏàòÎ∞ï','Î©îÎ°†','Îî∏Í∏∞','ÌÇ§ÏúÑ','Ïò§Î†åÏßÄ'],
    'col2' : [5000,3000,1000,600,700]
    }, index=[3,4,5,6,7])
```

```python
df1, df2
```

```python
pd.concat([df1, df2]) # ignore_index ÌååÎùºÎØ∏ÌÑ∞Îäî FalseÍ∞Ä ÎîîÌè¥Ìä∏Í∞íÏù¥Îã§. Í∏∞Ï°¥Ïùò Ïù∏Îç±Ïä§Î•º Ïú†ÏßÄÌïúÎã§.
# pd.concat([df1, df2], ignore_index=False, axis=0)ÏôÄ ÎèôÏùº
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1) # ÏπºÎüº Î†àÎ≤®Î°ú Î≥ëÌï©ÌïòÎäî Í≤ΩÏö∞, ÏÑúÎ°ú Îã§Î•∏ Ïù∏Îç±Ïä§ÏóêÏÑú Í≤∞Ï∏°ÏπòÍ∞Ä Î∞úÏÉùÌïúÎã§.
```

### 2) ÏπºÎüºÎ™ÖÏù¥ Îã§Î•∏ Í≤ΩÏö∞
concate() Î©îÏÜåÎìúÏùò ÌååÎùºÎØ∏ÌÑ∞ joinÏùÑ Ïù¥Ïö©ÌïúÎã§.
*   join=outer : Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Í∞Å ÏπºÎüºÎì§Ïùò Ìï©ÏßëÌï©
*   join=inner : Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Í∞Å ÏπºÎüºÎì§Ïùò ÍµêÏßëÌï©

```python
df3 = pd.DataFrame({
    'item':['item0','item1','item2', 'item3'],
    'count':['count0','count1','count2','count3'],
    'price':['price0','price1','price2','price3']
    }, index = [0,1,2,3])
df4 = pd.DataFrame({
    'item':['item2','item3','item4', 'item5'],
    'count':['count2','count3','count4','count5'],
    'price':['price2','price3','price4','price5'],
    'var':['var2','var3','var4','var5']
    }, index = [2,3,4,5])
```

```python
df3, df4
```

```python
pd.concat([df3, df4], join='outer') # ÏπºÎüºÎì§Ïùò Ìï©ÏßëÌï©
```

## 2 merge()ÏôÄ join()ÏùÑ Ïù¥Ïö©Ìï¥ Î≥ëÌï©ÌïòÍ∏∞
1. merge() :ÌäπÏ†ïÌïú keyÎ•º Í∏∞Ï§ÄÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Î≥ëÌï©Ìï† Ïàò ÏûàÎã§.
*   on='ÏπºÎüºÎ™Ö' : Ïñ¥Îñ§ ÏπºÎüºÏùÑ Í∏∞Ï§ÄÏúºÎ°ú mergeÌï†ÏßÄ ÏÑ§Ï†ï
*   how='ÏïÑÎûòÏùò Í∞íÎì§' : merge Î∞©ÏãùÏùÑ ÏÑ§Ï†ï
  * inner(default) : Í∏∞Ï§ÄÏúºÎ°ú ÏÇºÏùÄ ÏπºÎüº('ÏπºÎüºÎ™Ö')ÏóêÏÑú Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Î™®Îëê Ï°¥Ïû¨ÌïòÎäî Îç∞Ïù¥ÌÑ∞Îßå Í∞ÄÏ†∏Ïò®Îã§.
  * left : ÏôºÏ™Ω Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'ÏπºÎüºÎ™Ö'Ïùò Í∞íÎì§ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÌïúÎã§.
  * right : Ïò§Î•∏Ï™Ω Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'ÏπºÎüºÎ™Ö'Ïùò Í∞íÎì§ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú ÌïúÎã§.
  * outer : Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'ÏπºÎüºÎ™Ö'Ïùò Í∞íÎì§ÏùÑ Î™®Îëê ÏÇ¨Ïö©ÌïúÎã§.

```python
customer = pd.DataFrame({'customer_id' : np.arange(6),
                    'name' : ['James', 'Elly', 'Tom', 'Givert', 'Aiden', 'Brody'],
                    'ÎÇòÏù¥' : [40, 20, 21, 30, 31, 18]})

orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                    'item' : ['ÎßàÏö∞Ïä§', 'Ï∂©Ï†ÑÍ∏∞', 'Ïù¥Ïñ¥Ìè∞', 'Ìó§ÎìúÏÖã', 'Ï†ÑÏûêÌéú', 'ÌÇ§Î≥¥Îìú', 'Ï†ÑÏûêÌéú', 'ÎßàÏö∞Ïä§', 'ÌÇ§Î≥¥Îìú', 'ÏºÄÏù¥Ïä§'],
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
```

```python
pd.merge(customer, orders, on='customer_id')
# pd.merge(customer, orders, on='customer_id', how='inner')ÏôÄ ÎèôÏùºÌïòÎã§.
# customer_idÍ∞Ä Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê Î™®Îëê Ï°¥Ïû¨ÌïòÎäî Îç∞Ïù¥ÌÑ∞Îßå Í∞ÄÏ†∏Ïò®Îã§. => Í≤∞Ï∏°ÏπòÍ∞Ä Î∞úÏÉùÌïòÏßÄ ÏïäÎäîÎã§.
```

```python
pd.merge(customer, orders, on='customer_id', how='right') # Ïò§Î•∏Ï™Ω Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ Í∏∞Ï§ÄÏúºÎ°ú mergeÌïúÎã§. ‚≠ê Ïù¥Îïå ÏπºÎüºÏùò ÏàúÏÑúÎäî Î≥ÄÌïòÏßÄ ÏïäÎäîÎã§.
```

```python
pd.merge(customer, orders, on='customer_id', how='outer') # Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò 'customer_id'Î•º Î™®Îëê ÌôúÏö©Ìï¥ mergeÌïúÎã§.
```
