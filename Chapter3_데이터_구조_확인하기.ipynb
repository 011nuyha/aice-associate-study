## 2 데이터프레임의 기본 정보 확인하기
데이터프레임의 속성과 메소드를 활용하여 간단한 통계 정보를 확인할 수 있다.

```python
# shape는 (행의 개수, 열의 개수)를 튜플 형태로 반환
flight.shape
```

```python
# columns는 데이터프레임의 열 레이블을 반환
flight.columns
```

```python
# dtypes는 데이터프레임의 각 column의 데이터 타입을 반환
flight.dtypes
```

```python
# info() 메소드는 데이터 유형, 각 칼럼의 데이터 수 등을 한번에 볼 수 있음 => 데이터의 구성이나 결측치 유무를 확인 가능
# 결측치 : 데이터셋에서 값이 누락되거나 비어있는 상태
flight.info()
```

```python
# describe() 메소드는 기본적으로 int, float 같은 숫자형 데이터의 column에 대한 통계 정보를 제공한다.
# 범주형 데이터 정보(데이터 개수, 최빈값, 최빈값 개수)를 확인하려면 include='all'로 지정한다.
# std : 표준편차(standard deviation)
flight.describe()#include='all')
```

```python
# value_counts() 메소드는 지정된 열의 각 값(value)에 대한 발생 횟수를 보여준다.
flight['source_city'].value_counts()
```
